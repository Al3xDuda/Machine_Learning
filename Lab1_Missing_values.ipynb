```python
null_columns=['age','fare','cabin','embarked','boat','body','home_dest']
```

```python
def p_class_group(null_columns) :
    return dataset.groupby('pclass')[null_columns].apply(lambda x: x.isna().sum())
def survivor_group(null_columns) :
    return dataset.groupby('survived')[null_columns].apply(lambda x: x.isna().sum())
```

```python
p_class_group(null_columns)/dataset[null_columns].isna().sum()
```

We can clearly see what the author called it 'Third Class Bias' where most of the null values is in 3rd class. Most of 3rd class passangers were immigrants and travelling alone so probably nobody knew the exact details of their passage. Likely information about 3rd class passanger wasn't as important as the information about higher classes thus leading to negligence in data maintainance.

```python
survivor_group(null_columns)
```

```python
survivor_group(null_columns)/dataset[null_columns].isna().sum()
```

there is a less visibile 'survivor bias'. It was propably harder to find information on people that have not come back to shore. And thus one whould have to rely on any records existing before departure. Those could be badly maintained/lost therefore a visibile higer ratio of non-survived to survived number of nulls.

We are likely dealing here with MAR data since most missing values are in 3rd_class. I will impute age values as medians from each class and survivors. That is if a person has no age, is in the 2nd class and survived I will fill age with a median of all second class passengers that survived

```python
dataset.groupby(['pclass', 'survived'])['age'].median()
```

```python
dataset['age']=dataset['age'].fillna(dataset.groupby(['pclass', 'survived'])['age'].transform('median'))
```

As expected. The people that died were older on average.
